```python
import random

# Step 1: Generate random data
item_list = ["Apple", "Banana", "Carrot", "Tomato", "Pineapple", "Mango", "Watermelon", "Kiwi", "Grapes", "Guava"]
quantity_list = [random.randint(1, 10) for _ in range(100)]
price_list = [round(random.uniform(1.0, 10.0), 2) for _ in range(100)]

# Step 2: Create a CSV file manually
with open('sales_data.csv', 'w') as csvfile:
    csvfile.write("Item,Quantity,UnitPrice\n")
    for _ in range(100):
        random_index = random.randint(0, 9)
        item = item_list[random_index]
        quantity = quantity_list[_]
        price = price_list[_]
        csvfile.write(f"{item},{quantity},{price}\n")

# Step 3: Calculate and display total cost for each row
with open('sales_data.csv', 'r') as csvfile:
    csvfile.readline()  # Skip the header row
    print("Total Cost for the first 5 rows:")
    for i, line in enumerate(csvfile):
        if i < 5:
            item, quantity, price = line.strip().split(',')
            total_cost = float(quantity) * float(price)
            print(f"{item} | Quantity: {quantity} | Unit Price: \u20B9{price} | Total Cost: \u20B9{total_cost:.2f}")

# Step 4: Calculate and display total cost for each row
with open('sales_data.csv', 'r') as csvfile:
    csvfile.readline()  # Skip the header row
    print("Total Cost for the last 10 rows:")
    for i, line in enumerate(csvfile):
        if i > 89:
            item, quantity, price = line.strip().split(',')
            total_cost = float(quantity) * float(price)
            print(f"{item} | Quantity: {quantity} | Unit Price: \u20B9{price} | Total Cost: \u20B9{total_cost:.2f}")

# Step 4: Calculate total cost for each unique item
item_totals = {}
with open('sales_data.csv', 'r') as csvfile:
    csvfile.readline()  # Skip the header row
    for line in csvfile:
        item, quantity, price = line.strip().split(',')
        total_cost = float(quantity) * float(price)
        if item in item_totals:
            item_totals[item] += total_cost
        else:
            item_totals[item] = total_cost

print("\nTotal Cost for each unique item:")
print("Item Name | Total Quantity | Total")
for item, total_cost in item_totals.items():
    total_quantity = sum([float(quantity) for line in open('sales_data.csv') if item in line])
    print(f"{item} | {total_quantity} | \u20B9{total_cost:.2f}")

# Step 5: Print minimum and maximum prices for each item sold
item_prices = {}
with open('sales_data.csv', 'r') as csvfile:
    csvfile.readline()  # Skip the header row
    for line in csvfile:
        item, _, price = line.strip().split(',')
        price = float(price)
        if item in item_prices:
            item_prices[item].append(price)
        else:
            item_prices[item] = [price]

print("\nMinimum and Maximum Prices for Each Item Sold:")
for item, prices in item_prices.items():
    min_price = min(prices)
    max_price = max(prices)
    print(f"{item} | Min Price: \u20B9{min_price:.2f} | Max Price: \u20B9{max_price:.2f}")
```
