# Create a Python Class for managing student details and marks.
* Define the class and implement the methods of the student class in a menu-driven program for different types of users 

```python
class Student:
    def __init__(self, name, roll_number):
        self.name = name
        self.roll_number = roll_number
        self.marks = {}

    def add_marks(self, subject, marks):
        self.marks[subject] = marks

    def get_total_marks(self):
        return sum(self.marks.values())

    def get_average_marks(self):
        total_marks = self.get_total_marks()
        num_subjects = len(self.marks)
        if num_subjects > 0:
            return total_marks / num_subjects
        else:
            return 0

    def display_student_info(self):
        print(f"Name: {self.name}")
        print(f"Roll Number: {self.roll_number}")
        print("Marks:")
        for subject, marks in self.marks.items():
            print(f"{subject}: {marks}")
        print(f"Total Marks: {self.get_total_marks()}")
        print(f"Average Marks: {self.get_average_marks()}")

def main():
    students = {}

    while True:
        print("\nMenu:")
        print("1. Add a new student")
        print("2. Add marks for a student")
        print("3. Display student details")
        print("4. Quit")
        choice = input("Enter your choice: ")

        if choice == '1':
            name = input("Enter student name: ")
            roll_number = input("Enter roll number: ")
            student = Student(name, roll_number)
            students[roll_number] = student
            print("Student added successfully!")

        elif choice == '2':
            roll_number = input("Enter roll number of the student: ")
            if roll_number in students:
                subject = input("Enter subject: ")
                marks = float(input(f"Enter marks for {subject}: "))
                students[roll_number].add_marks(subject, marks)
                print("Marks added successfully!")
            else:
                print("Student not found. Please add the student first.")

        elif choice == '3':
            roll_number = input("Enter roll number of the student: ")
            if roll_number in students:
                students[roll_number].display_student_info()
            else:
                print("Student not found. Please add the student first.")

        elif choice == '4':
            print("Goodbye!")
            break

        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()

4
```
